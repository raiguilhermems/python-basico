## if, else, elif

diferença entre múltiplos if's e if + elif

* quando utilizo vários **if's**, eles são independentes, ou seja, o programa roda um por um e o **else** fica vinculado apenas ao último **if**

* quando utilizo **elif**, a estruta é única, de modo que atendendo uma das condições, ele sai da estrutura de comparação. Além disso, o **else** fica vinculado a estrutura toda e não apenas ao último **if**  
    **-geralmente as comparações são das mesmas variáveis**

* Lista.**append**(element) -> adiciona o elemento na última posição

## Removendo elementos da lista 

### Método remove

* Lista.**remove**(elemento) -> remove o **primeiro** elemento indicado da lista  

```
a = [4, 2, 3, 4, 5]  
a.remove(4)  
print(a)
>>> [2, 3, 4, 5]
```

**Perceba que apenas o primeiro 4 foi removido. Caso queira remover outro elemento (ou todos), devo utilizar outro método.*

### Método pop

* Lista.**pop**(elemento=[-1]) -> remove o **último** elemento OU o elemento informado como parâmetro.  

```
a = [4, 2, 3, 4, 5]  
a.pop()  
print(a)
>>> [4, 2, 3, 4]

a = [4, 2, 3, 4, 5]  
a.pop(2)  
print(a)
>>> [4, 2, 4, 5]
```

### Método copy

* lista.**copy**() -> copio os dados de uma lista para outra  
    *Quando faço:* 

```
a = [1, 2, 3]  
b = a

B fica dependente de A, e modificando A, modifico B (e vice-versa). Isso ocorre porque a Variável B na verdade não é criada, mas aponta para a alocação de memória de A: 

a = [1, 2, 3, 4]  
b = a  
a[0] = 100  
print(b)
>>> [100, 2, 3, 4]
```

* *A função copy faz com que a variável B seja realmente criada e independa de A. Isso é muito útil quando vou fazer listas auxiliares:*  

```
a = [1, 2, 3, 4]  
b = a.copy()  
a[0] = 100  
print(b)
>>> [1, 2, 3, 4]
```

## Definições

* **Classe** - um grupo de objetos com características semelhantes.

* **Instância / objeto** - um exemplar de uma classe.

* **Atributo** - são *características* de uma instâncias (seriam variáveis que armazem qualidades de um atributo).

* **Métodos** - são *ações* de uma instância (são funções que se aplicam aquela instância.

* **Herança** - mecanismo pelo qual uma sub-classe herda métodos e atributos da uma supler-classe.

* **Associação** - mecanismo pelo qual uma instância/objeto interage com uma instância de outra classe, utilizando recursos dessa segunda classe.

* **Polimorfismo** - quando por exemplo um mesmo método se comporta de forma diferente em sub-classes (ex: função calcula área de objetos diferentes).

* **Encapsulamento** - possibilidade de esconder atributos ou métodos, sendo acessiveis apenas de dentro da classe. Utilizo o __ (dunder) para esconder esse atributo/método.

## Herança

* Para fazer uma sub-classe, basta passar o nome da superclasse como parâmetro na criação da sub-classe:

```
class minhaSubClasse(Superclasse):
    pass
```

* Quando as duas classes tem método construtor, preciso chamar o construtor da superclasse, caso contrário não herdarei os atributos dessa classe pai.

```
classe Superclasse:
    def __init__(self, atributoA):
        self.atributoA = atributoA


classe minhaSubClasse(Superclasse):
    def __init__(self, atributoB, atributoA):
        super(minhaSubClasse, self).__init__(atributoA)
        self.atributoB = atributoB
```

Dessa forma, ao instanciar um objeto da subClasse, ele já pede como parâmetro os atributos A(superclasse) e B(subclasse).

Da mesma forma, (utilizando Super), posso invocar métodos da Superclasse na Subclasse.

## Atributos de classe - atributos estáticos

* São atributos declarados fora do método construtor e não pertencem as instâncias, mas sim a classe.

* Posso alterar esses atributos de dentro dos métodos das intâncias, utilizando:

```
minhaClasse.atributo = "novo valor"
```

Exemplo:
```
class Conta:
    total = 10000
    
    def __init__(self, id):
        self.id = id
    
    def deposito(self, valor):
        Conta.total += valor
```

Perceba que consigo acessar o atributo da classe utilizando:

```
Conta.total = "novo valor"
```

Também poderia modificar utilizando o método "depósito" com a **instância da classe**, e nesse caso passando alguma instância como parâmetro:

```
Conta.deposito(itau, 300)
```

Nesse exemplo itau é uma instância da classe Conta.  


Já quando utilizo o método por meio de uma instância da classe, o próprio Python passa a instância como self, não precisando passar esse argumento:

```
itau.deposito(300)
```

Ou seja, os seguintes comandos são equivalentes:
```
Conta.deposito(itau, 300)
itau.deposito(300)
```

# Erros e exceções

* Utilizo para evitar que os erros parem a execução do meu programa, além disso, utilizo para identificação e correção de erros ao longo do código.

```
try:  
    tenta executar esse bloco
    
except:
    se der erro executa essse bloco
    
else:
    é executado se não der nenhum erro
        
finally:
    executa esse bloco, de qualquer forma
   
```

* Ao utilizar o **try**, meu código executa *até encontrar um erro*, e caso encontre, pula para o **except**. Portanto, as linhas de código executadas antes de identificar o erro são mantidas.  

* Posso utilizar o **else** após esses vários excepts, **e só entra no bloco *else* se não entrar em nenhuma das exceções.**

## Especificando erros

* Posso especificar um tipo (ou vários tipos) de erro(s) para entrar no bloco except, exemplo:

```
except FileNotFoundError:

except (IndexError, ValueError, SyntaxError):
```

* Posso utilizar *alias* em exceções:

```
except (IndexError, ValueError, SyntaxError) as E:
    print(E)
```

## Levantando exceções

## Criando exceções

* Para criar uma exceção, crio uma classe que herda a superclasse Exception e utilizo o método mágico __ str__ para retornar a mensagem de erro. Utilizo o **raise** quando desejar levantar essa minha exceção.

```
class MinhaExcecao(Exception):

    def __str__(self) -> str:
        return 'Mensagem de erro da minha exceção'
        
raise MinhaExcecao
```

## Assert

```
def raizQuadrada(x):
    assert x>0, 'x não pode ser negativo'
    return x ** 1/2

raizQuadrada(-2)
```

## Statement with as

* Utilizo o statement with para lidar melhor com exceções ao manipular arquivos. Esse statement faz parte de um objeto **Context Manager**.

* É muito útil utilizar esse gerenciador de contexto porque não preciso me preocupar com fechar arquivo ou com erro de abrir arquivo inexiste (ele se encarrega de criar nesses casos).

```
with open('arquivo.txt', 'w') as file:
    file.write('Olá')
```    

### Objeto Context Manager

* Esse objeto possui dois métodos:

__ enter__ -> chamado ao entrar no gerenciador de contexto  

__ exit __ -> chamado ao encontrar um erro, retornando **tipo, valor e traceback**  
Quando não encontro erro, ele também passa pelo __ exit__ mas retorna *None* em cada um desses valores.



# Debugging com pdb

```

[expressão | for lop | clausula if(opcional)]

[x + 10 for x in range (10)]

________________________________

Equivalente a:

lista = []
for x in range(10):
    l.append(x + 10)

```

* A **expressão** é o valor que vou adicionar na minha lista, seguido da expressão tenho um loop for de um objeto iterável.


* Posso fazer um list comprehension com uma lista já existente, para filtra valors desejados:

```
l = minha lista original
l = [x for x in l if x%2 == 0]
```

No exemplo anterior filtrei os valores da lista l para manter só os valores pares.

## Nested loops list comprehension


```

l1 = [1, 2, 3]
l2 = [4, 5, 6]

list(zip(l1, l2))
>>> [(1, 4), (2, 5), (3, 6)]

dict(zip(l1, l2))
>>>{1: 4, 2: 5, 3: 6}
```

* Caso os objetos sejam de tamanhos diferentes, o zip para no primeiro erro (StopIteration). Portanto, o tamanho do objeto gerado depende do tamanho do menor objeto.


# Map

* A função map aplica uma função a cada um dos itens de um objeto iterável.

**map(func, ** iterators)**

* Muito semelhante a uma list comprehension:

```
l1 = [x**2 for i in range(4)]
>>> [0, 1, 4, 9]

l1 = map((lambda x : x**2), range(4))
>>> [0, 1, 4, 9]
```

Posso utilizar uma função diretamente ou um lambda gerado dentro da propria função map.

* É possível utilizar vários objetos iteráveis:

```
l1 = [1, 2, 3]
l2 = [4, 5, 6]
a = list(map((lambda x, y: x + y), l1, l2))
>>>[5, 7, 9]
```

No exemplo anterior, a função precisava de dois parâmetros, então l1 serviu como primeiro parâmetro e l2 como segundo.  

*A principal vantagem em relação ao list comprehension é a possibilidade de passar funções mais complexas.


# Filter 

* Recebe uma função e um objeto iterável.

**filter(func, iterable) -> filter objetct**  

* Para cada valor iterado, se o valor da função for True, ele retorna o valor para a lista.

```
a = list(filter((lambda x: x%2 = 0), range(5))
>>>[0, 2, 4]
```

No exemplo foi iterado os valores do range de 0 a 4 e caso o valor da função anônima fosse True, o valor era armazenado na lista.

* Utilizando map e filter combinados consigo uma maior versatilidade que as list comprehension:

```
a = list(map((lambda x: x**2), filter((lambda x: x%2 == 0), range(5))))
>>>[0, 4, 16]
_________________________________________________________
Equivalente a:

[x**2 for x in range(5) if x%2 == 0]
```


# All, any e sum

* all (iterable) -> verifica em um iterável se **todos** os elementos são *True

* any (iterable) -> verifica em um iterável se **pelo menos um** dos elementos é *True

* sum(iterable, start=0) -> soma valores númericos (*int* ou *float*) de um iterável e retorna o resultado. Posso passar um segundo argumento como inicio da minha soma:

```
sum([1.3, 2, 3], 10)
>>> 16.3
```

# Reduce

